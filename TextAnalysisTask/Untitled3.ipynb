```python
import re
import string

def cleaning(text):
    # تحويل النص إلى أحرف صغيرة وإزالة الروابط والشخصيات الخاصة وعلامات الترقيم
    text = text.lower()  # تحويل إلى أحرف صغيرة
    text = re.sub('https?://\S+|www\.\S+', '', text)  # إزالة الروابط
    text = re.sub(r"\b\d+\b", "", text)  # إزالة الأرقام
    text = re.sub('<.*?>+', '', text)  # إزالة الشخصيات الخاصة
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text)  # إزالة علامات الترقيم
    text = re.sub('\n', '', text)  # إزالة الأسطر الجديدة
    text = re.sub('[’“”…]', '', text)  # إزالة بعض الشخصيات الخاصة

    # إزالة الرموز التعبيرية:
    emoji_pattern = re.compile("["
                               u"\U0001F600-\U0001F64F"  # emoticons
                               u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                               u"\U0001F680-\U0001F6FF"  # transport & map symbols
                               u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                               u"\U00002702-\U000027B0"
                               u"\U000024C2-\U0001F251"
                               "]+", flags=re.UNICODE)
    text = emoji_pattern.sub(r'', text)

    # تحويل الاختصارات إلى نصوص كاملة:
    text = re.sub("isn't", 'is not', text)
    text = re.sub("he's", 'he is', text)
    text = re.sub("wasn't", 'was not', text)
    text = re.sub("there's", 'there is', text)
    text = re.sub("couldn't", 'could not', text)
    text = re.sub("won't", 'will not', text)
    text = re.sub("they're", 'they are', text)
    text = re.sub("she's", 'she is', text)
    text = re.sub("There's", 'there is', text)
    text = re.sub("wouldn't", 'would not', text)
    text = re.sub("haven't", 'have not', text)
    text = re.sub("That's", 'That is', text)
    text = re.sub("you've", 'you have', text)
    text = re.sub("He's", 'He is', text)
    text = re.sub("what's", 'what is', text)
    text = re.sub("weren't", 'were not', text)
    text = re.sub("we're", 'we are', text)
    text = re.sub("hasn't", 'has not', text)
    text = re.sub("you'd", 'you would', text)
    text = re.sub("shouldn't", 'should not', text)
    text = re.sub("let's", 'let us', text)
    text = re.sub("they've", 'they have', text)
    text = re.sub("You'll", 'You will', text)
    text = re.sub("i'm", 'i am', text)
    text = re.sub("we've", 'we have', text)
    text = re.sub("it's", 'it is', text)
    text = re.sub("don't", 'do not', text)
    text = re.sub("that´s", 'that is', text)
    text = re.sub("I´m", 'I am', text)
    text = re.sub("it’s", 'it is', text)
    text = re.sub("she´s", 'she is', text)
    text = re.sub("he’s'", 'he is', text)
    text = re.sub('I’m', 'I am', text)
    text = re.sub('I’d', 'I did', text)
    text = re.sub("he’s'", 'he is', text)
    text = re.sub('there’s', 'there is', text)

    return text

# تطبيق دالة التنظيف على عمود 'v2' (النصوص)
df['cleaned_text'] = df['review'].apply(cleaning)

# عرض بعض النتائج للتأكد من نجاح العملية
df[['review', 'cleaned_text']].head()
```

```python
# تحديد الكلمات الأكثر شيوعاً
FREQWORDS = set([w for (w, wc) in cnt.most_common(10)])

# دالة لإزالة الكلمات الأكثر شيوعاً
def remove_freqwords(text):
    """دالة مخصصة لإزالة الكلمات المتكررة"""
    return " ".join([word for word in str(text).split() if word not in FREQWORDS])

# تطبيق الدالة على عمود no_sw
df["wo_stopfreq"] = df['no_sw'].apply(lambda text: remove_freqwords(text))

# عرض البيانات بعد التعديل
df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from nltk.tokenize import RegexpTokenizer

# إعداد tokenizer و CountVectorizer
token = RegexpTokenizer(r'[a-zA-Z0-9]+')
cv = CountVectorizer(stop_words='english', ngram_range=(1,1), tokenizer=token.tokenize)

# Assuming 'review' or 'wo_stopfreq_lem' is the intended column instead of 'v2'
# Replace 'wo_stopfreq_lem' with the actual column name containing the text data
column_name = 'wo_stopfreq_lem'  # or 'review' if that's the correct column

# استبدال القيم الفارغة في العمود المحدد (النص)
df[column_name] = df[column_name].fillna('')

# تطبيق CountVectorizer على النصوص
text_counts = cv.fit_transform(df[column_name])

# عرض أسماء الميزات (الكلمات) وشكل المصفوفة الناتجة
words = cv.get_feature_names_out()
text_counts.shape, words[:10]  # عرض شكل المصفوفة وأول 10 كلمات
```

```python
from sklearn.model_selection import train_test_split

# X هو النصوص المعالجة بواسطة CountVectorizer
X = text_counts

# Assuming 'label' is the intended column name instead of 'v1'
# Replace 'label' with the actual column name containing the target variable
target_column = 'label'  # or another column name if 'label' is incorrect

# y هو العمود الذي يحتوي على التصنيفات (target_column)
y = df[target_column]

# تقسيم البيانات إلى مجموعة تدريبية واختبارية
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=30)
```

```python
from sklearn.naive_bayes import ComplementNB
from sklearn.metrics import classification_report, confusion_matrix
from sklearn import metrics
import pandas as pd

# تدريب نموذج ComplementNB على البيانات التدريبية
CNB = ComplementNB()
CNB.fit(X_train, y_train)

# التنبؤ باستخدام البيانات الاختبارية
predicted = CNB.predict(X_test)

# حساب دقة النموذج
accuracy_score = metrics.accuracy_score(predicted, y_test)

# طباعة نتائج التقييم
print('ComplementNB model accuracy is', str('{:04.2f}'.format(accuracy_score * 100)) + '%')
print('------------------------------------------------')
print('Confusion Matrix:')
print(pd.DataFrame(confusion_matrix(y_test, predicted)))
print('------------------------------------------------')
print('Classification Report:')
print(classification_report(y_test, predicted))
```
